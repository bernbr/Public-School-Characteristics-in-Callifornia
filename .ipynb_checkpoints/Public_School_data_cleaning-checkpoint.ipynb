```python
import pandas as pd
```

```python
# The path to our CSV file
file = "Public_School_Characteristics_2020-21_kw.csv"

# Read our Crowdfunding data into pandas
df = pd.read_csv(file)
df.head(10)
```

```python
# Get a list of all of our columns for easy reference
df.columns
```

```python
reduced_ps_df = df.loc[:, ["STATE_ID", "SCH_NAME", "LSTATE",
                            "LZIP", "VIRTUAL", "TOTFRL", "TOTAL", "PK", "KG",  
                            "G01", "G02", "G03", "G04", "G05",
                          "G06", "G07", "G08", "G09", "G10", 
                          "G10", "G11", "G12", "G13", "FTE", 
                           "CHARTER_TEXT", "MAGNET_TEXT", 
                           "LATCOD", "LONCOD"]]

reduced_ps_df

```

```python
# Create a new column that finds the average amount pledged to a project
reduced_ps_df.fillna(0, inplace=True)
reduced_ps_df.head()
```

```python
reduced_ps_df.dtypes
```

```python
reduced_ps_df["sum_elementary"]= reduced_ps_df[["PK","KG",'G01', 'G02', 'G03', 'G04', 'G05']].sum(axis=1)
```

```python
reduced_ps_df["sum_middle"]= reduced_ps_df[["G06","G07",'G08']].sum(axis=1)
```

```python
reduced_ps_df["sum_high"]= reduced_ps_df[["G09","G10",'G11', 'G12', 'G13']].sum(axis=1)
```

```python
reduced_ps_df.dtypes
```

```python
levels_ps_df = reduced_ps_df.loc[:, ["STATE_ID", "SCH_NAME", "LSTATE",
                            "LZIP", "VIRTUAL", "sum_elementary", "sum_middle", "sum_high",
                          "TOTFRL", "TOTAL", "FTE", 
                           "CHARTER_TEXT", "MAGNET_TEXT", 
                           "LATCOD", "LONCOD"]]
levels_ps_df.head()
```

```python
rename_ps_df = levels_ps_df.rename(columns={"STATE_ID": "Location_ID", 
                             "SCH_NAME": "School_Name", 
                             "LSTATE": "Location_Abbreviation",
                             "LZIP": "Location_Zip",
                             "VIRTUAL": "Virtual", 
                             "sum_elementary": "Elementary(PK-G5)",
                             "sum_middle": "Middle(G6-G8)", 
                             "sum_high": "High(G9-G13)",
                              "TOTFRL": "Total_Free/Reduced_Lunch", 
                             "TOTAL": "Total_Students", 
                             "FTE": "Total_Teachers", 
                           "CHARTER_TEXT": "Charter", 
                             "MAGNET_TEXT": "Magnet", 
                           "LATCOD": "Latitude",
                             "LONCOD": "Longitude"})
rename_ps_df.head()
```

```python
rename_ps_df.to_csv("Public_School_Characteristics_2020-21_cleaned.csv")
```
